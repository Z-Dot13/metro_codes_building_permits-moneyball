```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sw = pd.read_csv('../data/Initiated Building Permits (SW Tasks) 3yr.csv',
                 parse_dates=['Date_Initiated','Date_Issued','Task_Complete_Date'])
```

```python
sw.columns = sw.columns.str.lower()
```

```python
sw
```

```python
new_sw = sw[["permit #", "case_type", "case_type_desc", "status_code", "date_initiated", "date_issued", "task_code", "task_desc", "result_code", "result_desc", "task_complete_date"]]
new_sw
```

```python
new_sw.sort_values(by='permit #')
new_sw
```

```python
new_sw['deltatask'] = new_sw['task_complete_date'] - new_sw['date_initiated']
new_sw
```

```python
removed_t = new_sw[~new_sw['permit #'].str.startswith('T')]
removed_t
```
